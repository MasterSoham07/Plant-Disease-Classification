<h2>Importing Libraries</h2>

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

<h2>Data Preprocessing</h2>

<h3>Training Image Preprocessing</h3>

```python
training_set = tf.keras.utils.image_dataset_from_directory(
    'train',
    labels="inferred",
    label_mode="categorical",
    class_names=None,
    color_mode="rgb",
    batch_size=32,
    image_size=(128, 128),
    shuffle=True,
    seed=None,
    validation_split=None,
    subset=None,
    interpolation="bilinear",
    follow_links=False,
    crop_to_aspect_ratio=False,
)
```

<h3>Validation Image Preprocessing</h3>

```python
validation_set = tf.keras.utils.image_dataset_from_directory(
    'valid',
    labels="inferred",
    label_mode="categorical",
    class_names=None,
    color_mode="rgb",
    batch_size=32,
    image_size=(128, 128),
    shuffle=True,
    seed=None,
    validation_split=None,
    subset=None,
    interpolation="bilinear",
    follow_links=False,
    crop_to_aspect_ratio=False,
)
```

```python
training_set
```

```python
for x,y in training_set:
    print(x,x.shape)
    print(y,y.shape)
    break
```

<h2>Building Model</h2>

<h3>To Avoid Overshooting</h3>
1. Choose small learning rate default 0.001 we are taking 0.0001
<br>
2. There may be a chance of underfitting, so we increase the number of neurons
<br>
3. Add more Convolutional layers to extract more features from images there may be possiblilty that model is unable to capture relevant features or model is confused due to lack of features so feed with more features

```python
from tensorflow.keras.layers import Dense, Conv2D, MaxPool2D, Flatten, Dropout
from tensorflow.keras.models import Sequential
```

```python
model = Sequential()
```

<h3>Convolution, Maxpooling and Flatenning</h3>

```python
model.add(Conv2D(filters = 32, kernel_size = 3, padding= 'same', activation= 'relu', input_shape = [128,128,3]))
model.add(Conv2D(filters = 32, kernel_size = 3, activation = 'relu'))
model.add(MaxPool2D(pool_size = 2, strides = 2))
```

```python
model.add(Conv2D(filters = 64, kernel_size = 3, padding= 'same', activation= 'relu'))
model.add(Conv2D(filters = 64, kernel_size = 3, activation = 'relu'))
model.add(MaxPool2D(pool_size = 2, strides = 2))
```

```python
model.add(Conv2D(filters = 128, kernel_size = 3, padding= 'same', activation= 'relu'))
model.add(Conv2D(filters = 128, kernel_size = 3, activation = 'relu'))
model.add(MaxPool2D(pool_size = 2, strides = 2))
```

```python
model.add(Conv2D(filters = 256, kernel_size = 3, padding= 'same', activation= 'relu'))
model.add(Conv2D(filters = 256, kernel_size = 3, activation = 'relu'))
model.add(MaxPool2D(pool_size = 2, strides = 2))
```

```python
model.add(Conv2D(filters = 512, kernel_size = 3, padding= 'same', activation= 'relu'))
model.add(Conv2D(filters = 512, kernel_size = 3, activation = 'relu'))
model.add(MaxPool2D(pool_size = 2, strides = 2))
```

```python
model.add(Dropout(0.25)) # To avoid overfitting
```

```python
model.add(Flatten())
```

```python
model.add(Dense(units = 1500, activation = 'relu'))
```

```python
model.add(Dropout(0.4))
```

```python
# Output Layer
model.add(Dense(units = 38, activation = 'softmax'))
```

<h3>Compiling Model</h3>

```python
model.compile(optimizer = tf.keras.optimizers.Adam(
    learning_rate=0.0001), loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
model.summary()
```

<h2>Model Training</h2>

```python
training_history = model.fit(x = training_set, validation_data = validation_set, epochs = 10)
```

<h2>Model Evaluation</h2>

```python
# Model Evaluation on training set
train_loss, train_acc = model.evaluate(training_set)
```

```python
print("Training Accuracy: ", train_acc)
print("Training Loss: ", train_loss)
```

```python
# Model Evaluation on Validation set
val_loss, val_acc = model.evaluate(validation_set)
```

```python
print("Validation Accuracy: ", val_acc)
print("Validation Loss: ", val_loss)
```

<h2>Saving Model</h2>

```python
model.save('plant_disease_model.keras')
```

<h2>Accuracy Visualization</h2>

```python
epochs = [i for i in range(1,11)]
epochs
```

```python
plt.plot(epochs, training_history.history['accuracy'], color = 'red', label = 'Training Accuracy')
plt.plot(epochs, training_history.history['val_accuracy'], color = 'blue', label = 'Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

<h2>Some other metrics for model evaluation</h2>

```python
class_names = training_set.class_names
class_names
```

```python
test_set = tf.keras.utils.image_dataset_from_directory(
    'valid',
    labels="inferred",
    label_mode="categorical",
    class_names=None,
    color_mode="rgb",
    batch_size=32,
    image_size=(128, 128),
    shuffle=False,
    seed=None,
    validation_split=None,
    subset=None,
    interpolation="bilinear",
    follow_links=False,
    crop_to_aspect_ratio=False,
)
```

```python
y_pred = model.predict(test_set)
y_pred, y_pred.shape
```

```python
predicted_classes = tf.argmax(y_pred, axis=1)
predicted_classes
```

```python
true_categories = tf.concat([y for x,y in test_set], axis = 0)
true_categories
```

```python
Y_true = tf.argmax(true_categories, axis = 1)
Y_true
```

image.png

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(Y_true, predicted_classes, target_names = class_names))
```

```python
cm = confusion_matrix(Y_true, predicted_classes)
cm.shape
```

<h2>Confusion Matrix Visualization</h2>

```python
plt.figure(figsize = (40,40))
sns.heatmap(cm, annot = True, annot_kws = {'size' : 10})
plt.xlabel("Actual Class", fontsize = 30)
plt.ylabel("Predicted Class", fontsize = 30)
plt.title("Plant Disease Prediction Confusion Matrix", fontsize = 35)
plt.show()
```
